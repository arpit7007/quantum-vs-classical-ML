```python
# classical_models.py

import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score


def load_breast_cancer_processed():
    import numpy as onp

    train = onp.load("data/processed/bc_train.npz")
    val   = onp.load("data/processed/bc_val.npz")
    test  = onp.load("data/processed/bc_test.npz")

    X_train, y_train = train["x"], train["y"]
    X_val,   y_val   = val["x"],   val["y"]
    X_test,  y_test  = test["x"],  test["y"]

    return X_train, y_train, X_val, y_val, X_test, y_test


def train_logistic_regression():
    X_train, y_train, X_val, y_val, X_test, y_test = load_breast_cancer_processed()

    clf = LogisticRegression(max_iter=1000)
    clf.fit(X_train, y_train)

    y_val_pred = clf.predict(X_val)
    y_test_pred = clf.predict(X_test)

    print("Logistic Regression:")
    print("  Val accuracy:", accuracy_score(y_val, y_val_pred))
    print("  Test accuracy:", accuracy_score(y_test, y_test_pred))
    print("  Test F1:", f1_score(y_test, y_test_pred))
    print()
    return clf


def train_svm(kernel="rbf"):
    X_train, y_train, X_val, y_val, X_test, y_test = load_breast_cancer_processed()

    clf = SVC(kernel=kernel)
    clf.fit(X_train, y_train)

    y_val_pred = clf.predict(X_val)
    y_test_pred = clf.predict(X_test)

    print(f"SVM ({kernel} kernel):")
    print("  Val accuracy:", accuracy_score(y_val, y_val_pred))
    print("  Test accuracy:", accuracy_score(y_test, y_test_pred))
    print("  Test F1:", f1_score(y_test, y_test_pred))
    print()
    return clf


def train_mlp():
    """
    Simple MLP: input 4 features -> hidden layer (e.g. 8 neurons) -> output.
    """
    X_train, y_train, X_val, y_val, X_test, y_test = load_breast_cancer_processed()

    clf = MLPClassifier(
        hidden_layer_sizes=(8,),   # 1 hidden layer with 8 neurons
        activation="relu",
        max_iter=1000,
        random_state=0
    )
    clf.fit(X_train, y_train)

    y_val_pred = clf.predict(X_val)
    y_test_pred = clf.predict(X_test)

    print("MLP (1 hidden layer, 8 units):")
    print("  Val accuracy:", accuracy_score(y_val, y_val_pred))
    print("  Test accuracy:", accuracy_score(y_test, y_test_pred))
    print("  Test F1:", f1_score(y_test, y_test_pred))
    print()
    return clf


if __name__ == "__main__":
    train_logistic_regression()
    train_svm(kernel="linear")
    train_svm(kernel="rbf")
    train_mlp()
    
```
