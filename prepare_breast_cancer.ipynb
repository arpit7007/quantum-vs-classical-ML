```python
# prepare_breast_cancer.py

from sklearn.datasets import load_breast_cancer 
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA 
import os
import pickle
import numpy as np 

os.makedirs("data/processed", exist_ok=True)

data = load_breast_cancer()
X = data.data
y = data.target 

# ---------- split ----------
split = StratifiedShuffleSplit(n_splits=1, test_size=0.4, random_state=42)
for train_index, temp_index in split.split(X, y):
    X_train, X_temp = X[train_index], X[temp_index]
    y_train, y_temp = y[train_index], y[temp_index]

split_2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
for val_index, test_index in split_2.split(X_temp, y_temp):
    X_val, X_test = X_temp[val_index], X_temp[test_index]
    y_val, y_test = y_temp[val_index], y_temp[test_index]

# ---------- scaling ----------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_test_scaled  = scaler.transform(X_test)

# ---------- PCA (2 qubits -> 4 features) ----------
pca = PCA(n_components=4)
X_train_pca = pca.fit_transform(X_train_scaled)
X_val_pca   = pca.transform(X_val_scaled)
X_test_pca  = pca.transform(X_test_scaled)

# save scaler & pca
with open("data/processed/breast_cancer_scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

with open("data/processed/breast_cancer_pca.pkl", "wb") as f:
    pickle.dump(pca, f)

# save final processed arrays
np.savez("data/processed/bc_train.npz", x=X_train_pca, y=y_train)
np.savez("data/processed/bc_val.npz",   x=X_val_pca,   y=y_val)
np.savez("data/processed/bc_test.npz",  x=X_test_pca,  y=y_test)


```
