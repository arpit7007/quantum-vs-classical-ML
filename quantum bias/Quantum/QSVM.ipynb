```python
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

from qiskit.circuit.library import PauliFeatureMap
from qiskit_machine_learning.kernels import FidelityQuantumKernel

# -------------------------------
# Load prepared data (12 features)
# -------------------------------
X_train = np.load("../data/processed/X_train.npy")  # shape: (n_samples, 12)
X_test  = np.load("../data/processed/X_test.npy")
y_train = np.load("../data/processed/y_train.npy")
y_test  = np.load("../data/processed/y_test.npy")

# -------------------------------
# Quantum feature map
# 12 features → 4 qubits
# X, Y, Z rotations + entanglement
# -------------------------------
feature_map = PauliFeatureMap(
    feature_dimension=12,          # 12 classical features
    paulis=["X", "Y", "Z", "ZZ"],   # rotations on all 3 axes + entanglement
    reps=2,                         # keep reps low (very important)
    entanglement="full"
)

# -------------------------------
# Fidelity Quantum Kernel
# -------------------------------
quantum_kernel = FidelityQuantumKernel(
    feature_map=feature_map
)

# -------------------------------
# Compute kernel matrices
# -------------------------------
print("Computing quantum kernel matrices...")

K_train = quantum_kernel.evaluate(x_vec=X_train)
K_test  = quantum_kernel.evaluate(x_vec=X_test, y_vec=X_train)

# -------------------------------
# QSVM (precomputed kernel)
# -------------------------------
qsvm = SVC(kernel="precomputed")
qsvm.fit(K_train, y_train)

# -------------------------------
# Evaluation
# -------------------------------
y_pred = qsvm.predict(K_test)
acc = accuracy_score(y_test, y_pred)

print("Quantum QSVM Accuracy:",  acc)
```
