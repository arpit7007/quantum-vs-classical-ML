```python
import numpy as np
from sklearn.model_selection import train_test_split
import os

def generate_quantum_dataset(n_samples=500):
    np.random.seed(42)
    X = np.random.uniform(0, 2 * np.pi, size=(n_samples, 2))

    y = np.sign(np.sin(X[:, 0]) * np.sin(X[:, 1]))
    y = (y > 0).astype(int)

    return X, y


def prepare_and_save():
    X, y = generate_quantum_dataset()

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    os.makedirs("data/processed", exist_ok=True)

    np.save("data/processed/X_train.npy", X_train)
    np.save("data/processed/X_test.npy", X_test)
    np.save("data/processed/y_train.npy", y_train)
    np.save("data/processed/y_test.npy", y_test)

    print("✅ Quantum dataset prepared and saved successfully")


if __name__ == "__main__":
    prepare_and_save()
```
